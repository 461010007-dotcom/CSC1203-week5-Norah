```python
import math

radius = float(input("Enter the radius of a circle: "))

def calculate_area(r):
    return math.pi * (r ** 2)

calculate_circumference = lambda r: 2 * math.pi * r

area = calculate_area(radius)
circumference = calculate_circumference(radius)

print("Area:", round(area, 2))
print("Circumference:", round(circumference, 2))
```

```python
import re

user_input = input("Enter a string: ")
numbers_found = re.findall(r'\d+', user_input)
numbers_list = [int(num) for num in numbers_found]

def sum_all_numbers(lst):
    return sum(lst)

multiply_sum = lambda s: s * 2

total_sum = sum_all_numbers(numbers_list)
doubled_value = multiply_sum(total_sum)

print("Numbers found:", numbers_list)
print("Sum:", total_sum)
print("Doubled value:", doubled_value)
```

```python
import math

user_data = input("Enter a list of numbers separated by commas: ")
numbers_list = [float(x) for x in user_data.split(',')]

def calculate_squares(lst):
    square_lambda = lambda x: x ** 2
    return [square_lambda(n) for n in lst]

squares = calculate_squares(numbers_list)
maximum_square = max(squares)
sum_of_squares = sum(squares)
square_root_of_sum = math.sqrt(sum_of_squares)

print("List of squares:", squares)
print("Maximum square:", maximum_square)
print("Square root of the sum of squares:", square_root_of_sum)
```

```python
import re
import math

password = input("Enter a password: ")

has_upper = re.search(r'[A-Z]', password)
has_lower = re.search(r'[a-z]', password)
has_number = re.search(r'\d', password)

if has_upper and has_lower and has_number:
    print("Password strength: strong")
else:
    print("Password strength: weak")

scores_input = input("Enter 5 scores separated by spaces: ")
scores_list = [float(s) for s in scores_input.split()]

def calculate_average(lst):
    average_lambda = lambda x: sum(x) / len(x)
    return average_lambda(lst)

average_score = calculate_average(scores_list)
rounded_average = math.ceil(average_score)

print("Rounded average score:", rounded_average)
```

```python
import re
import math

paragraph = input("Enter a paragraph of text: ")
words_starting_with_vowel = re.findall(r'\b[aeiouAEIOU]\w*', paragraph)

def count_total_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

calculate_word_length = lambda word: len(word)
vowel_word_lengths = [calculate_word_length(w) for w in words_starting_with_vowel]

total_vowels = count_total_vowels(paragraph)
average_word_length = sum(vowel_word_lengths) / len(vowel_word_lengths) if vowel_word_lengths else 0

print("Words starting with a vowel:", words_starting_with_vowel)
print("Total vowels in paragraph:", total_vowels)
print("Average vowel-word length:", round(average_word_length, 2))
```

```python
numbers_list = []
for i in range(5):
    number = float(input("Enter a number: "))
    numbers_list.append(number)

list_sum = sum(numbers_list)
list_average = list_sum / len(numbers_list)
list_maximum = max(numbers_list)
list_minimum = min(numbers_list)

print("Sum:", list_sum)
print("Average:", list_average)
print("Maximum:", list_maximum)
print("Minimum:", list_minimum)
```

```python
student_scores = {}

for i in range(3):
    name = input("Enter student name: ")
    score = float(input("Enter student score: "))
    student_scores[name] = score

print("All students and scores:", student_scores)

highest_student = max(student_scores, key=student_scores.get)
print("Student with the highest score:", highest_student)
```

```python
input1 = input("Enter the first list of numbers (separated by space): ").split()
input2 = input("Enter the second list of numbers (separated by space): ").split()

set1 = set([int(x) for x in input1])
set2 = set([int(x) for x in input2])

print("Union:", set1.union(set2))
print("Intersection:", set1.intersection(set2))
print("Difference (Set 1 - Set 2):", set1.difference(set2))
```
